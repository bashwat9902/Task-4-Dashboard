```python
# STEP 1: Upload CSV file
from google.colab import files
uploaded = files.upload()

# STEP 2: Import libraries and load data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')

# STEP 3: Basic exploration
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())

# STEP 4: Univariate Analysis
# Age Distribution
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Distribution of Age')
plt.show()

# Fare Boxplot
sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare')
plt.show()

# Categorical - Survived Count
df['Survived'].value_counts().plot(kind='bar', title='Survival Count')
plt.show()

# Pclass Count
df['Pclass'].value_counts().plot(kind='bar', title='Passenger Class Count')
plt.show()

# STEP 5: Bivariate Analysis
# Survival by Age
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survived')
plt.show()

# Survival by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()

# Survival by Sex and Age
sns.violinplot(x='Sex', y='Age', hue='Survived', data=df, split=True)
plt.title('Survival by Sex and Age')
plt.show()

# STEP 6: Multivariate Analysis - Pairplot
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']].dropna(), hue='Survived')
plt.show()

# STEP 7: Multivariate Analysis - Heatmap (numeric only)
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# STEP 8: Summary Stats for Writing Insights
print("Survival Rate by Sex:")
print(df.groupby('Sex')['Survived'].mean(), '\n')

print("Survival Rate by Pclass:")
print(df.groupby('Pclass')['Survived'].mean(), '\n')

print("Average Fare by Survival:")
print(df.groupby('Survived')['Fare'].mean(), '\n')

print("Average Age by Survival:")
print(df.groupby('Survived')['Age'].mean(), '\n')
```
